```python
#load all the required python libraries for functions required for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in the data using the function below, as it is a csv file the white delimiter is NOT set to true
dataset = pd.read_csv('machine.data', names = ['vendor_name', 'model_name','MYCT','MMIN','MMAX', 'CACH','CHMIN','CHMAX','PRP','ERP'])
```

```python
#view the dataset to make sure it has been correctly loaded
dataset.head()
```

```python
#check the size of the dataframe created rows x columns
dataset.shape
```

```python
#check for any null values that may negatively influence future predictions
dataset.isnull().values.any()
```

```python
#view a basic summary of the dataset
dataset.describe()
```

```python
#view what kind of numbers or character are used to denote the characteristics of the dataframe
#as may be seen the values are in integer form and so transformations to convert them to numbers that can be used to work with are not required
dataset.dtypes
```

```python
#use the loaded data to plot histograms of the values of type int64
dataset.hist(figsize = (10,10))
#from the data it may be inferred that the data is skewed to the right (i.e. the tail is to the right)
#hence it may be said that a linear regression model is not appropriate for the data as it would not meet the normality assumption
#as a result a random forest method is used as an alternative. Additionally, a logarithmic transformation may be applied 
#then a linear regression model may be fitted to give a better representation of the data
```

```python
#a pairwise scatter plot is created describing the correlation between the variables
sns.pairplot(dataset, height = 2.0)
```

```python
#from this, it may be useful to generate a heatmap representing the value of the correlation between the data.
#as shown there are very weak positive correlations between many of the variables and a linear regression model may not be appropriate
ax, fig = plt.subplots(figsize = (16,10))
correlation_matrix = dataset.corr()
sns.heatmap(correlation_matrix, annot = True, cmap = "YlGnBu")
```

```python
#the ERP plot shows that the dependent variable is not exactly normally distributed
sns.distplot(dataset['ERP'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#after importing the required functions from said libraries the data is split into the dependent variable y = ERP
#and the independent variable x = all other integer values
x = dataset.drop(['ERP','vendor_name', 'model_name'], axis = 1)
y = dataset['ERP']
#show the initial output to make sure it has been stored
x.head()
```

```python
#using the train_test_split function the data is split into a training and test set to build the numerous classification trees which build the random forest
#the training data is then fitted into the random forest model
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size = 0.3, random_state = 42)
rfmodel = RandomForestClassifier(random_state = 42)
rfmodel.fit(train_x,train_y)
```

```python
#using score the 
predicted = rfmodel.predict(test_x)
rfmodel.score(test_x,test_y)
```

```python
#as can be seen the perfectly fitted data would have a graph idenitcal to y=x however there are major deviations on the scatter plot shown below
plt.scatter(test_y,predicted)
```

```python
#as such it may be stated that the model created is not good at predicting values and instead a different model perhaps
#a log transformed linear regression model may be more accurate in undergoing predictions
```

```python
#testing the distplot of a log transformed version of ERP i.e. the dependent variable
sns.distplot(dataset["ERP"].apply(np.log))
```

```python
#reshaping the data and dropping the vendors and model names
logdata = dataset.drop(['vendor_name', 'model_name'], axis = 1)
logdata.head()
```

```python
#undergoing a complete logarithmic transformation of all data to better fit a linear regression model
for column in logdata:
    logdata[column] = np.log(logdata[column] + 100)
logdata.head()
```

```python
#a pairwise scatter plot is created describing the correlation between the variables
sns.pairplot(logdata, size = 2.0)
```

```python
#round the values; then using the train_test_split function seperate it into distinct sets
logdata = logdata.round(5)
x = logdata.drop(['ERP'], axis = 1)
y = logdata['ERP']
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
#create a linear regression model and predict values of the test_x to get predicted y values and plot against true y values in the scatter plot
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_x,train_y)
predict = model.predict(test_x)
```

```python
plt.scatter(test_y,predict)
```

```python
#as shown this is a far better prediction of the dataset than the random forest performed earlier
model.score(test_x,test_y)
#the predicted value may be back transformed into the standard value by undergoing the calculation e^(number) - 100
#the +100 was added as a constant earlier to avoid np.inf values
```
